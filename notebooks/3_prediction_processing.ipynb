```python
import pandas as pd
import joblib
```

```python
df = pd.read_parquet('../data/processed/1_test_processed.parquet')

df
```

```python
clf = joblib.load('../models/clf_1.pkl')
hgbr = joblib.load('../models/hgbr_1.pkl')
```

```python
df_1 = df.copy()
```

```python
X = df_1.drop(columns=['id'])

df_1['is_zero'] = clf.predict(X)

df_1
```

```python
non_zero_rows = df_1[df_1['is_zero'] == False].copy()

non_zero_rows
```

```python
y = hgbr.predict(non_zero_rows.drop(columns=['is_zero', 'id']))

df_1.loc[df_1['is_zero'] == False, 'sales'] = y
df_1.loc[df_1['is_zero'] == True, 'sales'] = 0

df_1
```

```python
((len(df_1) - len(df_1[df_1.sales == 0])) / len(df_1)) * 100
```

```python
df_train = pd.read_csv('../data/row/train.csv')

((len(df_train) - len(df_train[df_train.sales == 0])) / len(df_train)) * 100
```

```python
df_1[['id', 'sales']].to_csv('../data/processed/submission_1.csv', index=False)
```
